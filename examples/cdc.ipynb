```python
import pandas as pd
import numpy as np
```

```python
url = 'https://www.cdc.gov/workplacehealthpromotion/data-surveillance/docs/whpps_120717.csv'
cdc = pd.read_csv(url, sep='~')
cdc.head()
```

## Filter columns

```python
cols_to_keep = ['Size', 'OC1', 'HPR9_1', 'HPR1_1A', 'HPR1_1', 'HS15_2', 'HPR2_3B', 'WD2']
cdc_clean = cdc[cols_to_keep]
cdc_clean
```

## Rename columns

```python
cdc_clean = cdc_clean.rename({'Size': 'size',
                              'OC1':'department',
                              'HPR9_1':'sleep',
                              'HPR1_1A':'physact',
                              'HPR1_1':'physprogram',
                              'HS15_2':'breastcancer',
                              'HPR2_3B':'vendfood',
                              'WD2':'percent_female'},
                            axis=1)
cdc_clean
```

## Recode and collapse categories

```python
cdc_clean['size'].value_counts()
```

```python
size_replace = {1.0: '10-24',
               2.0: '25-49',
               3.0: '50-99',
               4.0: '100-249',
               5.0: '250-499',
               6.0: '500-749',
               7.0: '750-999',
               8.0: '1000+'}
cdc_clean['size'] = cdc_clean['size'].replace(size_replace)
cdc_clean
```

```python
cdc_clean['vendfood'].value_counts()
```

```python
size_replace = {1.0:'10-24',
               2.0:'25-49',
               3.0:'50-99',
               4.0:'100-249',
               5.0:'250-499',
               6.0:'500-749',
               7.0:'750-999',
               8.0:'1,000+'}
cdc_clean['size'] = cdc_clean['size'].replace(size_replace)

 

department_replace = {'1':'Human Resources and/or Benefits', 
                      '2':'Health Promotion/Fitness/Wellness',
                      '3':'Safety',
                      '4':'Medical', 
                      '5':'PR/Marketing', 
                      '6':'Finance',
                      '7':'Other',
                      '97':np.nan,
                      '99':np.nan}

cdc_clean['department'] = cdc_clean['department'].astype('str').replace(department_replace)


vendfood_replace = {1.0: 'Yes',
                   2.0: 'No',
                   5.0: 'No',
                   96.0: np.nan,
                   97.0: np.nan,
                   98.0: np.nan,
                   99.0: np.nan}
cdc_clean['vendfood'] = cdc_clean['vendfood'].replace(vendfood_replace)

 

bc_replace = {1 : 'Yes',
                2: 'No',
                5 : 'No',
                9 : 'No',
                96 : np.nan,
                97 : np.nan,
                99 : np.nan}
cdc_clean['breastcancer'] = cdc_clean['breastcancer'].replace(bc_replace)

 


physact_replace = {1:'info',
               2:'Skillbuilding',
               3:'both',
               96:np.nan,
               97:np.nan,
               98:np.nan,
               99:np.nan}
cdc_clean['physact']=cdc_clean['physact'].replace(physact_replace) # The idea of replace is

 

sleep_replace = {1:'Yes',
                2:'No',
                5:'No',
                96:np.nan,
                97:np.nan,
                98:np.nan,
                99:np.nan}
cdc_clean['sleep'] = cdc_clean['sleep'].replace(sleep_replace)

 

physprograms_replace = {1.0:'Yes', 
                        2.0:'No', 
                        5.0:'No', 
                        99.0:np.nan, 
                        96.0:np.nan, 
                        97.0:np.nan, 
                        98.0:np.nan}
cdc_clean['physprogram'] = cdc_clean['physprogram'].replace(physprograms_replace)

percent_replace = {997:np.nan, 
                    998:np.nan, 
                    999:np.nan}

cdc_clean['percent_female'] = cdc_clean['percent_female'].replace(percent_replace)

cdc_clean
```
