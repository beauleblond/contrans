```python
import numpy as np
import pandas as pd
import os
import requests
import json
from bs4 import BeautifulSoup
import time
import collections
collections.Callable = collections.abc.Callable
```

# Goal: Get the text of all bills associated with Bob Good in the 11th Congress
## Step 1: Get Bob Good's ID number

```python
propublica_token = os.environ['propublica_token']

useragent_url = 'https://httpbin.org/user-agent'
r = requests.get(useragent_url)
useragent = json.loads(r.text)['user-agent']
```

```python
headers = {'X-API-Key': propublica_token,
          'User-Agent' : useragent,
          'From': 'bwl5cd@virginia.edu'}
```

```python
root = 'https://api.propublica.org'
congress = '117'
chamber = 'house'
endpoint = '/congress/v1/{congress}/{chamber}/members.json'.format(congress = congress,
                                                                      chamber = chamber)

r = requests.get(root + endpoint, headers=headers)
my_json = json.loads(r.text)
membersdf = pd.json_normalize(my_json, record_path = ['results', 'members'])
membersdf.head(3).T
```

```python
bobgood = membersdf.query("last_name == 'Good'")
```

```python
bobgood['id']
```

```python
bobgoodid = bobgood.reset_index()['id'][0]
```

## Step 2: Use BG's ID to query the bills API

```python
endpoint = '/congress/v1/members/{memberid}/bills/{billtype}.json'.format(memberid = bobgoodid, 
                                                                      billtype = 'introduced')
r = requests.get(root + endpoint, 
                headers = headers)
myjson = json.loads(r.text)
bgbills1 = pd.json_normalize(myjson, record_path= ['results', 'bills'])
```

```python
r = requests.get(root + endpoint, 
                headers = headers,
                params = {'offset':20})
myjson = json.loads(r.text)
bgbills2 = pd.json_normalize(myjson, record_path= ['results', 'bills'])
```

```python
bgbills = pd.concat([bgbills1, bgbills2], ignore_index=True)
bgbills.head(3).T
```

```python
def scrape_one_bill(url):
    time.sleep(2)
    print('Now getting the text from ' + url)
    r = requests.get(url,
                     headers = {'User-Agent': useragent,
                                'From': 'bwl5cd@virginia.edu'})
    my_html = BeautifulSoup(r.text, 'html.parser')
    try:
        billtext = my_html.find_all('pre')[0].text
        return billtext
    except:
        pass
```

```python
url_list = ['https://www.congress.gov/bill/117th-congress/house-bill/{i}/text?format=txt'.format(i=i) for i in range(3, 13)]
```

```python
bills = [scrape_one_bill(u) for u in url_list]
```
